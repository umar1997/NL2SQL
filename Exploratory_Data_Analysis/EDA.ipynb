### Knowledge Graph

```python
source, target, relation = [], [], []
for i, f in df.iterrows():
    relations_in_file = f['Relations']
    entities_in_file = f['Group_Entities']
    
    for rel in relations_in_file:
        ents = rel[2].split()
        entity_type = []
        for entity in entities_in_file:
            if (ents[0] == entity[0]) or (ents[1] == entity[0]):
                entity_type.append(entity[3])
        assert len(entity_type) == 2
    
#         print(rel[1], entity_type[0], entity_type[1])
        
        source.append(entity_type[0])
        target.append(entity_type[1])
        relation.append(rel[1])

assert len(source) == len(target) == len(relation)
```

```python
kg_df = pd.DataFrame({'source':source, 'target':target, 'edge':relation})
kg_df.head()
```

```python
G=nx.from_pandas_edgelist(kg_df, "source", "target", 
                          edge_attr=True, create_using=nx.MultiDiGraph())
```

```python
plt.figure(figsize=(12,12))

pos = nx.spring_layout(G)
nx.draw(G, with_labels=True, node_color='skyblue', edge_cmap=plt.cm.Blues, pos = pos)
plt.show()
```

```python
G=nx.from_pandas_edgelist(kg_df[kg_df['edge']=="Has_value"], "source", "target", 
                          edge_attr=True, create_using=nx.MultiDiGraph())

plt.figure(figsize=(12,12))
pos = nx.spring_layout(G, k = 0.5) # k regulates the distance between nodes
nx.draw(G, with_labels=True, node_color='skyblue', node_size=1500, edge_cmap=plt.cm.Blues, pos = pos)
plt.show()
```

```python
G=nx.from_pandas_edgelist(kg_df[kg_df['edge']=="OR"], "source", "target", 
                          edge_attr=True, create_using=nx.MultiDiGraph())

plt.figure(figsize=(12,12))
pos = nx.spring_layout(G, k = 0.5) # k regulates the distance between nodes
nx.draw(G, with_labels=True, node_color='skyblue', node_size=1500, edge_cmap=plt.cm.Blues, pos = pos)
plt.show()
```
